```python
import re

transcription = open('test/transcription.txt', 'r')
lines = transcription.readlines()

unformatted_transcripts = []
for line in lines:
    unformatted_transcripts.append(line.strip())


file1 = open("hinglish.txt", "r")
lines = file1.readlines()

hinglish_transcripts = []
for line in lines:
  #line = re.sub("\n", "", line)
  hinglish_transcripts.append(line.strip())
```

```python
formatted_transcripts = []

for transcript in unformatted_transcripts:
  trans = re.sub(r'\d+_\d+', '', transcript)
  
  formatted_transcripts.append(trans.strip())

print(len(formatted_transcripts))
formatted_transcripts_set = set(formatted_transcripts)
print(len(formatted_transcripts_set))
```

```python
formatted_transcripts_set
```

```python
hinglish_transcripts
```

```python
with open('hindi.txt', 'w') as f:
    for line in formatted_transcripts_set:
        f.write(f"{line}\n")
```

```python
# create list of 100
def split_list(lst, chunk_size):
    return [lst[i:i + chunk_size] for i in range(0, len(lst), chunk_size)]

chunk_size = 100
trans_batch = split_list(list(formatted_transcripts_set), chunk_size)
```

```python
len(trans_batch)
```

```python

from openai import OpenAI

open_ai_key = ''

client = OpenAI(
    api_key = open_ai_key,
)
```

```python
all_responces = []

for batch in trans_batch:
  completion = client.chat.completions.create(
  model="gpt-4o",
  messages=[
    {"role": "system", "content": "You are expert in writing Hinglish sentence. Write the sentence in Hinglish"},
    {"role": "user", "content": str(batch)}]
  )

  all_responces.append(completion)
```

```python
all_responces[0].choices[0].message.content
```

```python
batch_1 = all_responces[0].choices[0].message.content
batch_1 = re.sub(r'-', '', batch_1)

batch_1 = batch_1.split("\n")
batch_1 = [element.strip() for element in batch_1]

len(batch_1)
```

```python
all_responces[1].choices[0].message.content
```

```python
batch_2 = all_responces[1].choices[0].message.content
batch_2 = re.sub(r'\d+.', '', batch_2)

batch_2 = batch_2.split("\n")
batch_2 = [element.strip() for element in batch_2]

len(batch_2)
```

```python
all_responces[2].choices[0].message.content
```

```python
import ast

batch_3 = all_responces[2].choices[0].message.content
batch_3 = ast.literal_eval(batch_3)
batch_3 = [element.strip() for element in batch_3]
len(batch_3)
```

```python
all_responces[3].choices[0].message.content
```

```python
batch_4 = all_responces[3].choices[0].message.content
batch_4 = re.sub(r'\d+.', '', batch_4)

batch_4 = batch_4.split("\n")
batch_4 = [element.strip() for element in batch_4]

len(batch_4)
```

```python
len(formatted_transcripts_set)
```

```python
hinglish_list = []
hinglish_list.extend(batch_1)
hinglish_list.extend(batch_2)
hinglish_list.extend(batch_3)
hinglish_list.extend(batch_4)
len(hinglish_list)
```

```python
hinglish_dict = {}

for i in range(len(hinglish_list)):
  hinglish_dict[list(formatted_transcripts_set)[i]] = hinglish_list[i]
```

```python
hinglish_dict
```

```python
import pickle 

with open('hinglish_dict.pkl', 'wb') as f:
    pickle.dump(hinglish_dict, f)
```

```python
with open('hinglish_dict.pkl', 'rb') as f:
    hinglish_dict = pickle.load(f)
```
